## CUSTOMER SEGMENTATION USING K-MEANS CLUSTERING
K-means clustering is a technique used to organize data into groups based on their similarity.It is an Unsupervised Machine Learning algorithm which groups the unlabeled dataset into different clusters.

```python
# step1:normalizing the selected datas
from sklearn.preprocessing import StandardScaler
data_for_clus =['Income','Age','MntTotal','relatio_stats']
scaled_dataset = new_df.copy()
scaled_dataset[data_for_clus] =StandardScaler().fit_transform(new_df[data_for_clus])
scaled_dataset[data_for_clus].describe()
```

```python
# to get the optimum number of clusters we use silhouette score
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans

Y = scaled_dataset[data_for_clus]

silhouette_scores = []
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(Y)
    score = silhouette_score(Y, kmeans.labels_)
    silhouette_scores.append(score)
    lebels=kmeans.labels_

plt.figure(figsize=(12,8))
plt.plot(range(2, 11), silhouette_scores, marker='o')
plt.title('Silhouette Method')
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette Score')
plt.show()
```

The highest Silhouette score is for 4 clusters,there we will assign 4 clusters for the kmeans method

```python
print(lebels)
```

```python
model = KMeans(n_clusters=4, random_state = 42)
model.fit(scaled_dataset[data_for_clus])
scaled_dataset['lebels'] = model.predict(scaled_dataset[data_for_clus])
```

```python
new_df['lebels'] = scaled_dataset.lebels
new_df.groupby('lebels')[data_for_clus].mean()
```
